```python
import os

def check_model_path(path):
    if not os.path.exists(path):
        print(f"錯誤：模型路徑 {path} 不存在！")
        return False

    required_files = [
        "config.json",
        "pytorch_model.bin",
        "text_encoder/config.json"
    ]

    missing_files = []
    for f in required_files:
        full_path = os.path.join(path, f)
        if not os.path.exists(full_path):
            missing_files.append(f)

    if missing_files:
        print(f"錯誤：模型資料夾缺少以下檔案：")
        for mf in missing_files:
            print(f" - {mf}")
        return False

    print(f"模型路徑 {path} 檢查通過！")
    return True

# 範例用法，請改成你的模型路徑
model_path = "/content/animatediff/models/RealisticVision"
check_model_path(model_path)
```
